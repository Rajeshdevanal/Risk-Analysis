```python
import matplotlib.pyplot as plt

cat_col = df[["nationality","occupation","fee_structure","income_band",
              "amount_of_credit_cards","properties_owned",
              "loyalty_classification","risk_weighting",
              "brid","genderid","iaid"]].columns

for col in cat_col:
    
    print(f"\nValue count of '{col}':\n")
    print(df[col].value_counts())
    
    plt.figure(figsize=(6,4))
    df[col].value_counts().plot(kind="bar")
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()   
```

## Bivariate Analysis
- Analysis of two variables to determine their relationship.
- Identify relationships between variables
- Detect correlations
- Understand patterns or trends
- Support decision-making and modeling

```python
import seaborn as sns
for col in cat_col:   
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x=col, hue="genderid")
    plt.show()
```

## Univariate Analysis
 - execute  at one column
 - Analysis of a single variable
 - Understand the data distribution 
 - Find outliers
 -  Check frequency or counts

```python
for col in cat_col:   
    plt.figure(figsize=(8,5))
    sns.countplot(data=df, x=col)
    plt.show()
```

## Multivariate Analysis
- more than two variables at the same time
- analyzing more than two variables at the same time to understand relationships and patterns among them.
- Understand complex relationships
- Identify patterns between multiple variables
- Detect interactions between variables
- Improve machine learning models

```python
for col in cat_col:
    if col != "genderid":
        
        plt.figure(figsize=(8,5))
        sns.countplot(data=df, x=col, hue="genderid")
        
        plt.title(f"{col} vs Gender")
        plt.xticks(rotation=45)
        
        plt.show()
```

```python
# Bivariate Analysis
for i, predictor in enumerate(cat_col):
    plt.figure(i)
    sns.countplot(data=df, x=predictor, hue='nationality')
```

```python
# HIstplot of value counts for different Occupation

for col in cat_col:
  if col == "occupation":
    continue
  plt.figure(figsize=(8,5))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

## Numerical Analysis

```python
no_col = ['estimated_income', 'superannuation_savings', 'credit_card_balance', 'bank_loans', 'bank_deposits', 'checking_accounts', 'saving_accounts', 'foreign_currency_account', 'business_lending']

# Univariate analysis and visualization
plt.figure(figsize=(25,25))
for i,col in enumerate(no_col):
  plt.subplot(6, 2, i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
plt.show()
```

## Heatmaps

```python
cor_rel_matrix = df[no_col].corr()

plt.figure(figsize=(12,12))
sns.heatmap(cor_rel_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

## Insights of EDA :

1. The correlation analysis reveals that bank deposits have strong relationships with checking accounts and saving accounts, indicating that customers with higher deposits maintain higher balances across multiple account types. Additionally, moderate correlations between credit card balances, loans, and business lending suggest that customers utilizing one financial service are likely to engage with multiple banking products.
